```python
import numpy as np
import cv2
from tensorflow.keras.models import load_model
from matplotlib import pyplot

def save_plot(x,examples, n):
    examples = (examples + 1) / 2.0
    for i in range(n * n):
        pyplot.plot(n, n, i+1)
        pyplot.axis("off")
        pyplot.imshow(examples[i, :,:,0], cmap='gray')
    filename = f"/content/drive/MyDrive/OCT Project/OCT Code & Dataset/OCT Gan/gaan_drusen_1_iteration_3/output_1/sample2-{x+1}.png"
  
    pyplot.savefig(filename,bbox_inches='tight', pad_inches = 0)
    pyplot.close()

 
for x in range(1500):
    if __name__ == "__main__":
        model = load_model("/content/drive/MyDrive/OCT Project/OCT Code & Dataset/OCT Gan/gaan_drusen_1_iteration_3/Output/g_model.h5")

        n_samples = 1     ## n should always be a square of an integer.
        latent_dim = 128
        latent_points = np.random.normal(size=(n_samples, latent_dim))
        examples = model.predict(latent_points)
        save_plot(x,examples, int(np.sqrt(n_samples)))
else:
  print("Finally finished!")
```
